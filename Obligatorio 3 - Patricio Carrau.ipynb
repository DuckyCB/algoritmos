## Ejercicio 1
Probar que la recurrencia $T(n) = 2T(\frac{n}{2}) + n = Θ(nlgn)$ probando $T(n) = Ω(n lg n)$ (la cota
superior está probada en el libro)<ul>


Sabemos que $f(n)=\Theta(g(n)) \Leftrightarrow f(n)=O(g(n))$ $\&$ $f(n)=\Omega(g(n))$.

Ya está probado que $f(n)=O(n\lg(n))$ se cumple, por lo que buscamos probar $f(n)=\Omega(n\lg(n))$.

Para probar que se cumple $T(n)=\Omega (n\lg(n))$, suponemos $T(n)\geq c n \lg(n)$ para todo $n\geq n_0$, con $c$ y $n_0$ constantes positivas.

Comenzamos asumiendo que las condiciones se cumplen para un $m$ tal que $m<n$, particularmente para $m=\frac{n}{2}$.

Hipotesis: $T(n)\geq cn\lg(n)$

Tesis: $T(\frac{n}{2}) \geq c\frac{n}{2}\lg(\frac{n}{2})$

Tenemos entonces:

$T(n)=2T(\frac{n}{2})+n$ 

$\geq 2c\frac{n}{2} \lg(\frac{n}{2}) + n = 2c\frac{n}{2} (\lg(n)-\lg(2)) + n = cn\lg(n) - cn\lg(2)+n = cn\lg(n) - cn + n = cn\lg(n)+(1-c)n$

Con $c\leq 1$ obtenemos:

$T(\frac{n}{2}) \geq c\frac{n}{2}\lg(\frac{n}{2})$



## Ejercicio 2

Podemos ver que el árbol cuenta con dos ramas: Una con un crecimiento de $\frac{n}{3}$ y la otra con un crecimiento de $\frac{2n}{3}$.

La cota inferior $\Omega$ está determinada por la rama que termina antes. Como $\frac{n}{3}\leq \frac{2n}{3}$, sabemos que $\frac{n}{3}$ terminará primero, por lo que es la rama relacionada con $\Omega$.

De aqui podemos tomar que los nodos son $(\frac{1}{3})^kn$, de donde

$(\frac{1}{3})^kn=1$ $\Rightarrow$ $n=3^k$, de donde obtenemos $\lg_3n$.

Tenemos entonces:

$T(n)=\sum\limits^{\lg_3(n)}_{i=0}cn$

$= (\lg_3(n)+1)cn = cn\lg_3n+cn = \frac{cn\lg(n)}{\lg_3(n)}+cn\geq cn\lg(n)$ $\forall c\geq \frac{\lg_3}{n}$ 

$= \Omega(n\lg(n))$

Ya que encontramos una cota inferior de la función
